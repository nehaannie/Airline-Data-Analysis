**2.How the number of tickers booked and the amount earned changed with time?**

```python
tickets_and_bookings = pd.read_sql_query(''' 
                    select * from tickets 
                    join bookings 
                    on tickets.book_ref = bookings.book_ref
                    order by book_date asc
                    ''', connection)
```

```python
tickets_and_bookings["book_date"] = pd.to_datetime(tickets_and_bookings["book_date"])
tickets_and_bookings["date"] = tickets_and_bookings["book_date"].dt.date
tickets_and_bookings
```

```python
x = tickets_and_bookings.groupby("date")[["date"]].count()
plt.figure(figsize=(15,5))
plt.plot(x.index, x["date"], marker="p")
plt.xlabel("Date")
plt.ylabel("Tickets")
plt.grid('b')
plt.show()
```

**5.Average occupancy per aircraft**

```python
boarding_passes.head()
```

```python
flights.head()
```

```python
pd.read_sql_query('''
select aircraft_code, flights.flight_id, count(*) as seat_counts from boarding_passes
join flights on
boarding_passes.flight_id = flights.flight_id
group by aircraft_code, flights.flight_id
''', connection)
```

```python
pd.read_sql_query("select aircraft_code, count(*) as num_seats from seats group by aircraft_code;", connection)
```

```python
#Merging both tables
occupancy_rate = pd.read_sql_query('''
select a.aircraft_code, avg(a.seat_counts) as booked_seats, b.num_seats, avg(a.seat_counts)/b.num_seats as occupancy_rate from
(select aircraft_code, flights.flight_id, count(*) as seat_counts from boarding_passes
join flights on
boarding_passes.flight_id = flights.flight_id
group by aircraft_code, flights.flight_id) 
as a
inner join
(select aircraft_code, count(*) as num_seats from seats group by aircraft_code) 
as b
on a.aircraft_code = b.aircraft_code group by a.aircraft_code;
''', connection)
```

```python
occupancy_rate['inc_occupancy_rate'] = occupancy_rate['occupancy_rate'] + occupancy_rate['occupancy_rate'] * 0.1
```

```python
occupancy_rate
```

```python
total_revenue = pd.read_sql_query('''
SELECT aircraft_code, sum(amount) as total_revenue FROM ticket_flights
JOIN flights
ON ticket_flights.flight_id = flights.flight_id
GROUP BY aircraft_code
''', connection)
total_revenue
```

```python
occupancy_rate
```

```python
occupancy_rate["inc_total_annual_turnover"] = (total_revenue["total_revenue"]/occupancy_rate["occupancy_rate"]) * occupancy_rate["inc_occupancy_rate"]
```

**7. Which cities have the most airports?**

```python
city_airports = pd.read_sql_query('''
    SELECT
        json_extract(city, '$.en') AS city_name,
        COUNT(*) AS airport_count
    FROM airports_data
    GROUP BY city_name
    ORDER BY airport_count DESC
    LIMIT 10;
''', connection)
```

```python
plt.figure(figsize=(15,5))
plt.bar(city_airports['city_name'], city_airports['airport_count'], color='blue')
plt.xlabel("City")
plt.ylabel("Number of Airports")
plt.title("Top 10 Cities with Most Airports")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**8.What are the most common time zones for airports?**

```python
timezone_distribution = pd.read_sql_query('''
    SELECT
        timezone,
        COUNT(*) AS airport_count
    FROM airports_data
    GROUP BY timezone
    ORDER BY airport_count DESC
    LIMIT 10;
''', connection)
```

```python
plt.figure(figsize=(15,5))
plt.bar(timezone_distribution['timezone'], timezone_distribution['airport_count'], color='green')
plt.xlabel("Time Zone")
plt.ylabel("Number of Airports")
plt.title("Top 10 Most Common Time Zones for Airports")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**9.What are the top aircraft models by range?**

```python
top_aircrafts_by_range = pd.read_sql_query('''
    SELECT
        json_extract(model, '$.en') AS aircraft_model,
        range
    FROM aircrafts_data
    ORDER BY range DESC
    LIMIT 10;
''', connection)
```

```python
plt.figure(figsize=(15,5))
plt.bar(top_aircrafts_by_range['aircraft_model'], top_aircrafts_by_range['range'], color='skyblue')
plt.xlabel("Aircraft Model")
plt.ylabel("Range (km)")
plt.title("Top 10 Aircraft Models by Range")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**10.Which seats are the most frequently occupied?**

```python
popular_seats = pd.read_sql_query('''
    SELECT
        seat_no,
        COUNT(*) AS frequency
    FROM boarding_passes
    GROUP BY seat_no
    ORDER BY frequency DESC
    LIMIT 10;
''', connection)
```

```python
plt.figure(figsize=(10,5))
plt.bar(popular_seats['seat_no'], popular_seats['frequency'], color='green')
plt.xlabel("Seat Number")
plt.ylabel("Frequency")
plt.title("Top 10 Most Frequently Occupied Seats")
plt.grid(True)
plt.show()
```

**11.Which fare condition generates the most revenue overall?**

```python
total_revenue_by_fare = pd.read_sql_query('''
    SELECT
        fare_conditions,
        SUM(amount) AS total_revenue
    FROM ticket_flights
    GROUP BY fare_conditions
    ORDER BY total_revenue DESC;
''', connection)
```

```python
plt.figure(figsize=(10,5))
plt.bar(total_revenue_by_fare['fare_conditions'], total_revenue_by_fare['total_revenue'], color='orange')
plt.xlabel("Fare Conditions")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Fare Conditions")
plt.grid(True)
plt.show()
```
